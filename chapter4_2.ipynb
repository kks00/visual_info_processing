```python
import numpy as np
import cv2
from show_image import show_rgb_image
from show_image import show_images
%matplotlib inline
```

```python
#색상 값 정의
blue, green, red = ((255, 0, 0), (0, 255, 0), (0, 0, 255))
```

---

- ### 직선, 사각형 그리기

- ### cv2.line(image, pt1, pt2, color, thickness, lineType, ...)
    - #### thickness: 선 두께
    - #### lineType
        - #### cv2.LINE_4, cv2.LINE_8
        - #### cv2.LINE_AA: 안티앨리어싱(계단 현상 제거)

- #### (50, 50) 에서 (250, 150) 까지 두께 2인 선 그리기

```python
image = np.zeros((300, 300, 3), dtype="uint8")
image[:] = 255

cv2.line(image, (50, 50), (250, 150), red, 2)

show_images({"": image}, 5)
```

<br><br>

- ### cv2.rectangle
    - #### (image, pt1, pt2, color, thickness, lineType)
        - ##### pt1: 왼쪽 상단, pt2: 오른쪽 하단
    - #### (image, roi, color, thickness, lineType)
        - ##### roi: (왼쪽 상단 x, 왼쪽 상단 y, 가로 길이, 세로 길이)
    - #### thickness
        - ##### cv2.FILLED(-1): 내부 채우기

```python
image = np.zeros((400, 400, 3), dtype="uint8")
image[:] = 255

cv2.rectangle(image, (50, 50), (250, 150), blue, 3, cv2.LINE_AA)

cv2.rectangle(image, (50, 200, 200, 100), red, 3, cv2.LINE_AA)

cv2.rectangle(image, (300, 200, 50, 50), green, cv2.FILLED)

show_images({"": image}, 5)
```

---

- ### 글자 쓰기

- ### cv2.putText
    - #### (image, text, org, fontFace, fontScale, color, thickness, lineType, ...)
        - ##### org: 글자의 왼쪽 상단 좌표 (x, y)
        - ##### fontFace: cv2.FONT_HERSHEAY_SIMPLEX ...

```python
image = np.zeros((500, 600, 3), dtype="uint8")
image[:] = 255

cv2.putText(image, "SIMPLEX", (100, 100), cv2.FONT_HERSHEY_SIMPLEX, 2, red)
cv2.putText(image, "DUPLEX", (100, 200), cv2.FONT_HERSHEY_DUPLEX, 3, green)
cv2.putText(image, "ITALIC", (100, 300), cv2.FONT_HERSHEY_PLAIN | cv2.FONT_ITALIC, 4, blue)

show_images({"": image}, 8)
```

---

- ### 원 그리기

- ### cv2.circle
    - #### (img, center, radius, color, thickness, lineType, ...)
        - ##### lineType: cv2.FILLED(-1) 채우기

```python
image = np.zeros((500, 300, 3), dtype="uint8")
image[:] = 255

cv2.circle(image, (100, 100), 80, green, 2)
cv2.circle(image, (150, 300), 100, green, -1)

show_images({"": image}, 7)
```

<br><br>

- ### cv2.ellipse
    - #### (image, center, axes, angle, startAngle, endAngle, color, thickness, lineType, ...)
        - ##### axes: (가로, 세로) 반지름 길이
        - ##### angle: 회전 각도

```python
image = np.zeros((300, 900, 3), dtype="uint8")
image[:] = 255

cv2.ellipse(image, (100, 150), (120, 60), 90, 30, 270, red, 1)
cv2.ellipse(image, (300, 150), (120, 60), 270, -90, 180, red, 1)
cv2.ellipse(image, (500, 150), (120, 60), -45, 45, 270, red, 1)
cv2.ellipse(image, (700, 150), (120, 60), 90, -45, 90, red, 1)

show_images({"": image}, 8)
```
