```python
import numpy as np
import cv2
from show_image import show_rgb_image
from show_image import show_images
%matplotlib inline
```

```python
image = np.full((500, 500), 255, dtype="uint8")

tri_start = (100, 100)
tri_r = 200

pos_a = tri_start
pos_b = (tri_start[0], tri_start[1] + tri_r)
pos_c = (tri_start[0] + tri_r, tri_start[1] + tri_r)

cv2.putText(image, "pos_a", pos_a, cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 0))
cv2.putText(image, "pos_b", pos_b, cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 0))
cv2.putText(image, "pos_c", pos_c, cv2.FONT_HERSHEY_SIMPLEX, 1, (0, 0, 0))
            
cv2.line(image, pos_a, pos_b, (0, 0, 0), 2, cv2.LINE_AA)
cv2.line(image, pos_b, pos_c, (0, 0, 0), 2, cv2.LINE_AA)
cv2.line(image, pos_a, pos_c, (0, 0, 0), 2, cv2.LINE_AA)

show_images({"": image}, 5)
show_images({"flip 0": cv2.flip(image, 0), "flip 1": cv2.flip(image, 1), "flip -1": cv2.flip(image, -1)}, 10)
show_images({"transpose": cv2.transpose(image)}, 10)
```
